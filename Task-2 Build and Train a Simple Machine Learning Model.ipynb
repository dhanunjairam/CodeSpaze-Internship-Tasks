```python
#Loading the dataset 
import pandas as pd 
data = pd.read_csv("/Users/ABC/Downloads/insurance.csv")
data.head()
```

```python
data.info() #viewing the info os dataset
```

```python
data.isnull().sum() #to know about null values
```

```python
from sklearn.preprocessing import LabelEncoder,StandardScaler #imported labelencoder and standardscaler
encoder = LabelEncoder()
data['sex'] = encoder.fit_transform(data['sex'])
data['smoker'] = encoder.fit_transform(data['smoker'])
data['region'] = encoder.fit_transform(data['region'])
```

```python
data.head()#data after label encoding 
```

```python
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data[['bmi', 'charges']])  
```

```python
x = data.drop(['charges'],axis=1) #loading remaining features into x except tagert feature
y = data['charges'] #target feature
```

```python
from sklearn.model_selection import train_test_split #importing train test split module to split the data 
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=0) 
```

```python
from sklearn.linear_model import LinearRegression #importing linearregression model
model = LinearRegression() 
model.fit(X_train,y_train) #fitting the model with train data
```

```python
#defining a function to know the performance of the models
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score #
import numpy as np
def print_error(X_test, y_test, model_name):
  prediction = model_name.predict(X_test)
  print('Mean Absolute Error:', mean_absolute_error(y_test, prediction))
  print('Mean Squared Error:', mean_squared_error(y_test, prediction))
  print('Root Mean Squared Error:', np.sqrt(mean_squared_error(y_test, prediction)))
  print('R2_score:', r2_score(y_test, prediction))
```

```python
print_error(X_test,y_test,model) 
```

```python
from sklearn.tree import DecisionTreeRegressor #importing DecisionTreeRegressor 
dtr = DecisionTreeRegressor()
dtr.fit(X_train,y_train)
```

```python
print_error(X_test,y_test,dtr)
```

```python
#noe cross validatio
from sklearn.model_selection import KFold, cross_val_score
kf = KFold(n_splits=3,shuffle=True,random_state=42)

score = cross_val_score(dtr,x,y,scoring='neg_mean_squared_error',cv=kf)
score1 = cross_val_score(model,x,y,scoring='neg_mean_squared_error',cv=kf)
mse_scores = -score
mse_scores1 = -score1
```
